```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 2*np.random.rand(100,1)
y = 4+3*x+np.random.rand(100,1)

x1 = np.linspace(0,2,100)
y1 = 4.5+3*x1
```

```python
plt.scatter(x,y, color = 'b') #การใช้ scatter เวลาบอก สีให้ใส่ color ด้วย
plt.plot(x1,y1,'r')
plt.xlabel('$x$',fontsize = 18)
plt.ylabel('$y$',rotation = 0 ,fontsize = 18)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
model.fit(x,y) #ถูก fit แล้ว
```

```python
model.score(x,y) #modelนี้ถ้าค่าออกมาเท่ากับ 1 คือ  fit เลย 
```

```python
model.intercept_ # theta0
```

```python
model.coef_ #theta1
```

```python
y2 = model.intercept_+model.coef_*x1
```

```python
y2 = y2.reshape(100,1)
```

```python
x1=x1.reshape(100,1)
```

```python
plt.scatter(x,y)
plt.plot(x1,y2,'r')
plt.plot(x1,y1,'y')
plt.show()
```
