```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
#from matplotlib import pyplot as plt
```

```python
#ในคำสั่ง plot จะให้ใส่ค่าของ x y 
style.use('ggplot')
x = [5,8,10]
y = [12,16,6]
x2 = [6,9,11]
y2 = [5,15,7]
plt.plot(x,y,'b',label = 'line1')
plt.plot(x2,y2,'r', label = 'line2',marker = 'o',linewidth = 3)
plt.title('Topic info')
plt.ylabel('y axis')
plt.xlabel('x axis')
#legend คือการแสดงชื่อของเส้นกราฟที่กำหนดไว้
plt.legend()
plt.grid(True,color ='k')
plt.show()
```

```python
#ในคำสั่ง plot จะให้ใส่ค่าของ x y 
style.use('fivethirtyeight')
x = np.linspace(6,10,50)
y = 8*np.sin(x)
x2 = [6,9,11]
y2 = [5,15,7]
plt.plot(x,y,label = 'line1')
plt.plot(x2,y2,label = 'line2',marker = 'o',linewidth = 3)
plt.title('Topic info')
plt.ylabel('y axis')
plt.xlabel('x axis')
#legend คือการแสดงชื่อของเส้นกราฟที่กำหนดไว้
plt.legend()
plt.grid()
plt.show()
```

```python
#Bar plot
import matplotlib.pyplot as plt 

plt.bar([1,3,5,7,9],[5,2,7,8,2],label = "Example one")
plt.bar([2,4,6,8,10],[8,6,2,5,6],label = "Example two",color='g')
plt.legend()
plt.xlabel("bar number")
plt.ylabel("bar height")
plt.title("Info")
plt.show()
```

```python
#histogram
population_ages = [22,55,62,45,21,22,34,42,42,4,99,102,110,120,121,122
                   ,130,111,115,112,80,75,6,5,54,44,43,42,48]
bins = list(range(0,140,10))

plt.hist(population_ages, bins, histtype='bar',rwidth=0.2)
plt.xlabel('x')
plt.ylabel('y')
plt.title("histogram")
plt.legend()
plt.show()
```

```python
#scatter plot
style.use('fivethirtyeight')
x= [1,2,3,4,5,6,7,8]
y= [5,2,4,2,1,4,5,2]

plt.scatter(x,y, label='skitscat',color = 'k')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Scatter plot")
plt.grid()
plt.legend()
plt.show()
```

```python
#Stack plot
days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot(days,sleeping,color= 'm', label='Sleeping', linewidth = 5)
plt.plot(days,eating,color= 'c', label='Eating', linewidth = 5)
plt.plot(days,working,color= 'r', label='Working', linewidth = 5)
plt.plot(days,playing, color='k', label='Playing', linewidth = 5)

plt.stackplot(days,sleeping,eating,working,playing, colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title("Stack Plot")
plt.legend()
plt.show()

```

```python
#Pie plot
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']

plt.pie(slices,
       labels = activities,
       colors=cols,
       startangle=90,
       shadow=True,
       explode=(0,0.1,0,0),
       autopct='%.1f%%')
plt.title('Pie Plot')
plt.show()
```

```python
#Subplot
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0,5,0.1)
t2 = np.arange(0,5,0.02)

plt.subplot(121)
plt.plot(t1,f(t1),'bo',t2,f(t2))

plt.subplot(122)
plt.plot(t2, np.cos(2*np.pi*t2))

plt.show()
```
