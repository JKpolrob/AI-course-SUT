```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 2*np.random.rand(100,1)
```

```python
y = 4+3*x+np.random.rand(100,1)
```

```python
x1 = np.linspace(0,2,100)
y1 = 4.5+3*x1
```

```python
plt.scatter(x,y, color = 'b') #การใช้ scatter เวลาบอก สีให้ใส่ color ด้วย
plt.plot(x1,y1,'r')
plt.xlabel('$x$',fontsize = 18)
plt.ylabel('$y$',rotation = 0 ,fontsize = 18)
plt.show()
```

```python
lr = 0.02              #learning rate
max_iteration = 100
theta = np.ones([2,1]) #theta0,theta1
m = len(y)             # Number of Data
x_b = np.c_[np.ones([len(x),1]),x]
```

```python
theta
```

```python
x_b
```

```python
for i in range(max_iteration):
    predicted_y = np.dot(x_b,theta)
    #update eq
    theta = theta -(1/m)*lr*(x_b.T.dot((predicted_y-y)))
    print('Iteration {} : Theta0 = {:,.3f},Theta1 = {:,.3f}'.format(i+1,theta[0,0],theta[1,0]))
```

```python
x2 = np.linspace(0,2,100)
y2 = theta[0,0]+theta[1,0]*x2

plt.scatter(x,y)
plt.plot(x1,y1,'r')
plt.plot(x2,y2,'y.')
plt.show()
```
