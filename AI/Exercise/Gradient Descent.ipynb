# แบบที่ 1 มีการกำหนด precision และ จำนวน iteration max 

```python
cur_x = 3              #ค่า x ปัจจุบัน
rate = 0.01            #Learning rate
precision = 0.000001   #ค่า error ทีี่ยอมรับได้
previous_step_size = 1 #ค่า error ปัจจุบัน
max_iters = 10000      #จำนวนรอบการคำนวณ maximum number of iteration
iters = 0              #iteration conter current
df = lambda x:2*(x+5)  #Gradient of function
```

```python
while previous_step_size>precision and iters<max_iters:
    prev_x = cur_x
    cur_x = prev_x  - rate*df(prev_x )
    previous_step_size = abs(cur_x-prev_x)
    iters += 1
    print("Iteration",iters,"\nX value is",cur_x)

print("The local minimum occurs at", cur_x)
```

# แบบที่2 ใส่ค่ารอบที่ต้องการ ไม่ได้กำหนก precision

```python
x_previous = 3
lr = 0.01
max_iters =595

def J(x):
    return (x+5)**2

def dJ(x):
    return 2*(x+5)
```

```python
for i in range(max_iters):
    x_new = x_previous - lr*dJ(x_previous)
    x_previous = x_new
    print("Iteration {}, x_new = {}".format(i+1,x_new))
```

```python
import numpy as np
import matplotlib.pyplot as plt

J = lambda x: (x+5)**2    
x = np.linspace(-10,5,500)

def dJ(x):
    return 2*(x+5)  
def Gradient_descent(x_random,precision,l_r,error=1):
    
    x_list, y_list = [x_random], [J(x_random)]
    
    while error > precision:
        x_prev = x_random
        x_random = x_prev - (l_r * dJ(x_prev))
        error = abs(x_random - x_prev)
        x_list.append(x_random)
        y_list.append(J(x_random))
        
    print ("Different x = ",abs(x_random - x_prev))
    print ("final x     = ",x_random)
    print ("Local minimum occurs at: "+str(x_random))
    print ("Number of steps: " +str(len(x_list)))
    
    #plt.subplot(1,2,2)
    plt.scatter(x_list,y_list,c="b")
    plt.scatter(x_list[-1],y_list[-1],c="r",label = 'Local minimum occurs='+ str(x_random))
    plt.plot(x,J(x), c="y")
    plt.title("Gradient descent")
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.show()
```

```python
#Implement gradient descent (all the arguments are arbitrarily chosen)
Gradient_descent(1,0.00001,0.05)
```
