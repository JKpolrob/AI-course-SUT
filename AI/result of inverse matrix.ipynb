```python
import numpy as np
from numpy.linalg import inv
Matrix = df.values
M1 = Matrix[0:3,0].reshape(-1,1)
M2 = Matrix[0:3,1].reshape(-1,1)
M3 = Matrix[0:3,2].reshape(-1,1)
M4  = Matrix[0:3,3].reshape(-1,1)
Y = M4
MS = np.concatenate([M1,M2,M3], axis = 1)
InvMS = inv(MS)
#W = np.linalg.solve(MS.T*MS,MS.T*Y)
print(Y)
print(InvMS)
def Matrix_R(InvMS,Y):
#fine result of matrix
    Sum = 0
    ans = []
    for i in range(len(InvMS)):
        for j in range(len(Y)):
            Sum += InvMS[i,j]*Y[j,0]
        ans.append([Sum])
        Sum = 0 
    return np.array(ans)

a = inv(np.array([[1, -1, 1,2,1],[1,1,1,1,1],[1,3,9,9,10],[1,2,3,1,2],[4,5,6,1,2]]))
b = np.array([[-1],[1],[27],[1],[2]])

Matrix_R(InvMS,Y)#.astype(np.float32)
Matrix_R(a,b)#.astype(np.float32)

print(a)
a[:,0]
```
