```python
import numpy as np
import matplotlib.pyplot as plt
import math

t = np.linspace(-5, 1, 100)
front = np.exp(-t)
back  = np.cos(2*math.pi*t)
front_diff = -np.exp(-t)
back_diff = -2*math.pi*np.sin(2*math.pi*t)

diff_f_values = front*back_diff+back*front_diff

#ทำการ plot x เทียบ y ทั้ง 2 เส้นโดยเส้นแรกมีสีแดงเป็นเส้นทึบ เสน้ที่สองเป็นสีน้ำเงินเส้นประ
plt.plot(t, diff_f_values, 'r')
#ทำการแบ่งขอบเขตของกราฟที่จะแสดงสองช่องแรกเป็น xเริ่ม ถึง xสุดท้าย และ yเริ่ม ถึง yสุดท้าย
plt.axis([-5,1,-800, 550])
#saveภาพและตั้งชื่อไฟล์
plt.savefig('f.jpg')
#แสดงผล
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

t = np.linspace(-3, 2, 5000)
g_values = np.sin(t)*np.cos(1/t)

#ทำการ plot x เทียบ y ทั้ง 2 เส้นโดยเส้นแรกมีสีแดงเป็นเส้นทึบ เสน้ที่สองเป็นสีน้ำเงินเส้นประ
plt.plot(t, g_values, 'g')
#ทำการแบ่งขอบเขตของกราฟที่จะแสดงสองช่องแรกเป็น xเริ่ม ถึง xสุดท้าย และ yเริ่ม ถึง yสุดท้าย
plt.axis([-0.2,0.2,-0.2, 0.18])
#saveภาพและตั้งชื่อไฟล์
plt.savefig('f.jpg')
#แสดงผล
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib import style

def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

def df(t):
    front = np.exp(-t)
    back  = np.cos(2*math.pi*t)
    front_diff = -np.exp(-t)
    back_diff = -2*math.pi*np.sin(2*math.pi*t)
    diff_f_values = front*back_diff+back*front_diff
    return diff_f_values

def g(t):
    g = np.sin(t)*np.cos(1/t)
    return g

style.use('fivethirtyeight')

tf = np.linspace(-5,1,100)
tg = np.linspace(-3,2,5000)

plt.figure()
plt.subplot(221)
plt.plot(tf,f(tf),color = 'b')
plt.grid(True)
plt.title('f of exp(-t)cos(2*pi*t)')
plt.axis([-5,1,-100, 100])

plt.subplot(222)
plt.plot(tf,df(tf),color = 'r')
plt.grid(True)
plt.title('the derivation of f')
plt.axis([-5,1,-800, 600])

plt.subplot(223)
plt.plot(tg,g(tg),color = 'g')
plt.grid(True)
plt.title('The function g of sin(t)*cos(1/(t))')

plt.subplot(224)
plt.plot(tg,g(tg),color = 'k')
plt.grid(True)
plt.axis([-0.2,0.2,-0.2, 0.18])
plt.title('A closer look at g')

plt.subplots_adjust(0,-0.5,1.5,0.97,0.4,0.4)
plt.show()
```
